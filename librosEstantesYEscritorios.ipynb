# **TDA Pila:**

```python


# TDA
class Pila:
  def __init__(self,entrada = None):                         # Constructor
    self.pila = []
    if entrada:
      for dato in entrada:
        self.pila.append(dato)
  
  def __repr__(self):
    return str(self.pila)

  def vaciar(self):                                          # Vacia la pila
    self.pila.clear()

  def estaVacia(self):
    return len(self.pila) == 0                               # Pila vacia

  def apilar(self, item):                                    # Suma un item a la pila
    self.pila.append(item)

  def desapilar(self):                                       # Retira el ultimo elemento de la pila (si hay)
    dato = None
    if not self.estaVacia():
      dato = self.pila.pop()
    return dato

  def primerElemento(self):                                  # Devuelve el primer elemento de la pila (si hay)
    if self.estaVacia():
      print("La lista se encuentra vacia.")
      return None
    return self.pila[-1]

  def clonar(self):                                         # Devuelve una copia de la pila
    pilaClon = Pila()
    for elemento in self.pila:
      pilaClon.pila.append(elemento) #### pilaClon.apilar(elemento)
    return pilaClon

  def tamano(self):                                          # Devuelve el tamaño de la pila
    #print("El tamaño de la pila es:")
    return len(self.pila)

  def estaVacia(self):                                       # Devuelve TRUE si la pila se encuentra vacia
    return len(self.pila) == 0

  # Funciones Auxiliares
  def invertirPila(self):
    pilaAux = self.clonar()
    self.vaciar()
    while not pilaAux.estaVacia():
      self.apilar(pilaAux.desapilar())

```

# **TDA Estanterias**

```python
# Validaciones
def validaNumID(idDeEstanteria):
  if 0 < idDeEstanteria < 999:
    return idDeEstanteria
  else:
    raise Exception("El ID ingresado excede los limites.")


# TDA
class Estanteria: 
  
  def __init__(self,numID = int):
    self.pilaDeAutoresNacionales = Pila()
    self.pilaDeAutoresInternacionales = Pila()   # Constructor
    self.numID = validaNumID(numID)
                          
  def __repr__(self):
    return "Numero de ID:" + " " + str(self.numID) + str(self.pilaDeAutoresNacionales) + str(self.pilaDeAutoresInternacionales)

  def guardarLibro(self,libro):
    cantCritica = 50
    if libro.nacionalidad == "nacional" and self.pilaDeAutoresNacionales.tamano() < cantCritica: 
      self.pilaDeAutoresNacionales.apilar(libro)
    if libro.nacionalidad == "internacional" and self.pilaDeAutoresInternacionales.tamano() < cantCritica:
      self.pilaDeAutoresInternacionales.apilar(libro)

  def pilasDeAutores(self):
    return self.pilaDeAutoresNacionales, self.pilaDeAutoresInternacionales

  def primerLibroDisponible(self):
    primerElemento = None
    if not self.pilaDeAutoresNacionales.estaVacia():
      primerElemento = self.pilaDeAutoresNacionales.primerElemento()
    elif not self.pilaDeAutoresInternacionales.estaVacia():
       primerElemento = self.pilaDeAutoresInternacionales.primerElemento()
    else:
      primerElemento = "La estanteria esta vacia."
    return primerElemento  

  def libroParaRecomendar(self,generoDeLibro):
    libroARecomendar = None
    auxNacional = self.pilaDeAutoresNacionales.clonar()
    auxInternacional = self.pilaDeAutoresInternacionales.clonar()

    while not auxNacional.estaVacia() :
      if auxNacional.primerElemento().genero == generoDeLibro:
         libroARecomendar = auxNacional.primerElemento()
      auxNacional.desapilar()   

    if libroARecomendar == None:
      while not auxInternacional.estaVacia():
        if auxInternacional.primerElemento().genero == generoDeLibro:
          libroARecomendar = auxInternacional.primerElemento()
        auxInternacional.desapilar()

    if  libroARecomendar == None:
      raise Exception("No hay un libro que se pueda recomendar en estos momentos.")

    return libroARecomendar      
  
  def buscarLibro(self,codigoLibro):
    libroBuscado = None
    nacionalesAux = self.pilaDeAutoresNacionales.clonar()
    internacionalesAux = self.pilaDeAutoresInternacionales.clonar()
    
    while not nacionalesAux.estaVacia():
      if nacionalesAux.primerElemento().codigoDeIdentificacion != codigoLibro:
          libroBuscado = nacionalesAux.primerElemento()
      nacionalesAux.desapilar()
        
    # Lo mismo pero con la pila Internacional
    if  libroBuscado == None:
      while not internacionalesAux.estaVacia():
        if internacionalesAux.primerElemento().codigoDeIdentificacion != codigoLibro:
          libroBuscado = internacionalesAux.primerElemento()
        libroBuscado = nacionalesAux.desapilar()
    
    return libroBuscado

  def prestarLibro(self,codigoLibro):  
    nacionalesAux = self.pilaDeAutoresNacionales.clonar()
    nacionalesAux.invertirPila()
    self.pilaDeAutoresNacionales.vaciar()
    libroParaMostrar = None

    while not nacionalesAux.estaVacia():
      if nacionalesAux.primerElemento().codigoDeIdentificacion != codigoLibro:
        self.pilaDeAutoresNacionales.apilar(nacionalesAux.desapilar())
      else:
        libroParaMostrar = nacionalesAux.desapilar()

    # Lo mismo pero con la pila Internacional
    internacionalesAux = self.pilaDeAutoresInternacionales.clonar()
    internacionalesAux.invertirPila()
    self.pilaDeAutoresInternacionales.vaciar()
    while not internacionalesAux.estaVacia():
      if internacionalesAux.primerElemento().codigoDeIdentificacion != codigoLibro:
        self.pilaDeAutoresInternacionales.apilar(internacionalesAux.desapilar())
      else:
        libroParaMostrar = internacionalesAux.desapilar()
        
    return libroParaMostrar

  def librosPorTipo(self):
    return self.tamanoDePilaDeAutoresNacionales(),self.tamanoDePilaDeAutoresInternacionales()
  
  def cantidadTotalLibros(self):
    return self.tamanoDePilaDeAutoresNacionales() + self.tamanoDePilaDeAutoresInternacionales()

  def esCritica(self):
    return self.tamanoDePilaDeAutoresNacionales() > 50 or self.tamanoDePilaDeAutoresInternacionales() > 50 # Se reemplazo el tamaño critico por 2 para simplificar procesos
    
  def librosPorGenero(self,generoLibro):
    totalLibrosPorGenero = 0
    nacionalesAux = self.pilaDeAutoresNacionales.clonar()
    invertirPila(nacionalesAux)
    self.pilaDeAutoresNacionales.vaciar()

    while not nacionalesAux.estaVacia():
      if nacionalesAux.primerElemento().genero != generoLibro:
        self.pilaDeAutoresNacionales.apilar(nacionalesAux.desapilar())
      else:
        totalLibrosPorGenero += 1
        self.pilaDeAutoresNacionales.apilar(nacionalesAux.desapilar())

    # Lo mismo pero con la pila Internacional
    internacionalesAux = self.pilaDeAutoresInternacionales.clonar()
    invertirPila(internacionalesAux)
    self.pilaDeAutoresInternacionales.vaciar()
    
    while not internacionalesAux.estaVacia():
      if internacionalesAux.primerElemento().genero != generoLibro:
        self.pilaDeAutoresInternacionales.apilar(internacionalesAux.desapilar())
      else:
        totalLibrosPorGenero += 1
        self.pilaDeAutoresInternacionales.apilar(internacionalesAux.desapilar())
    
    return totalLibrosPorGenero 


  # Funciones auxiliares
  def tamanoDePilaDeAutoresNacionales(self):
    return self.pilaDeAutoresNacionales.tamano()

  def tamanoDePilaDeAutoresInternacionales(self):
    return self.pilaDeAutoresInternacionales.tamano()

  
# Pruebas

```
